```python
# 파일에서 2번째 컬럼에 있는 값을 붙이기
lst = []
with open(r'D:\corpus_sample\info\2BT_004.txt', 'r', encoding='cp949') as f:
    for line in f:
        line = line.rstrip()
        infos = line.split('\t')
        raw_ecel = infos[1]
        lst.append(raw_ecel)

txt = ' '.join(lst)

# 문장 단위로 분리하기
sents = txt.split('.')

with open('corpus_ex.txt', 'w', encoding='utf8') as f:
    for sent in sents:
        f.write(sent.lstrip()+'.\n')
```

```python
# 파일에서 3번째 컬럼에 있는 값을 붙이기
lst = []
with open(r'D:\corpus_sample\info\2BT_004.txt', 'r', encoding='cp949') as f:
    for line in f:
        line = line.rstrip()
        infos = line.split('\t')
        tagged_ecel = infos[2]
        lst.append(tagged_ecel)

txt = ' '.join(lst)

with open('corpus_ex.txt', 'w', encoding='utf8') as f:
    f.write(txt)
```

```python
freq_d = {}

# 형태소 빈도 산출
with open(r'D:\corpus_sample\info\2BT_004.txt', 'r', encoding='cp949') as f:
    for line in f:
        line = line.rstrip()
        infos = line.split('\t')
        
        tagged_ecel = infos[2]
        for mt in tagged_ecel.split('+'):
            mt = mt.strip()
            if mt not in freq_d:
                freq_d[mt] = 0
            freq_d[mt] += 1
```

```python
for mt, freq in freq_d.items():
    # N V M I
    if '/N' in mt or '/V' in mt or '/M' in mt or '/I' in mt:
        print(f'{mt.split("/")[0]}: {freq}')
```

```python
freq_d = {}

# 품사태그 빈도 산출
with open(r'D:\corpus_sample\info\2BT_004.txt', 'r', encoding='cp949') as f:
    for line in f:
        line = line.rstrip()
        infos = line.split('\t')
        tagged_ecel = infos[2]
        for mt in tagged_ecel.split('+'):
            mt = mt.strip()
            tag = mt.split('/')[-1]
            if tag not in freq_d:
                freq_d[tag] = 0
            freq_d[tag] += 1
```
