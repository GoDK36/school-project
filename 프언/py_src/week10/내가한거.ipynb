```python
from textblob import TextBlob
import nltk
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
#영문 텍스트 불러오기
with open(r"D:\GOD's folder\자료\대학교\언어학\프로그래밍 연습\교수님 자료\debate.txt", 'r', encoding='utf8') as inF:
    txt = inF.read().lower()
```

```python
stop = stopwords.words('english')
```

```python
stop
```

```python
blob = TextBlob(txt)
```

```python
ctt_words = []
for w in blob.words.singularize():
    if w in stop:
        continue
    ctt_words.append(w)
```

```python
ctt_words_nltk =nltk.Text(ctt_words)
```

```python
ctt_words_nltk.plot(30)
```

```python
#전체 단어 대상
blob_all = TextBlob(txt)
txt_nltk = nltk.Text(blob_all.words)
```

```python
blob_all.words
```

```python
print(txt_nltk.concordance('to', width = 80, lines=600))
#width는 to를 기준으로 양 쪽 80개, lines는 600개의 결과 이내로 보여줘라 라는 뜻
```

```python
txt_nltk.dispersion_plot(['trump', 'clinton'])
#텍스트에 어느부분에 나타나는지 보여줌
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
```

```python
with open(r"D:\GOD's folder\자료\대학교\언어학\프로그래밍 연습\교수님 자료\debate.txt", 'r', encoding='utf8') as inF:
    txt = inF.read().lower()
```

```python
stops = stopwords.words('english')
```

```python
blob = TextBlob(txt)
ctt_words = []
for w in blob.words.singularize():
    if w in stops:
        continue
    ctt_words.append(w)
print(ctt_words[:10])
```

```python
#빈도 사전 만들기
freq_d = Counter(ctt_words)
list(freq_d.items())[:5]
```

```python
#워드클라우드 생성 코드
wordcloud = WordCloud()
worldcloud = wordcloud.generate_from_frequencies(freq_d)
```
