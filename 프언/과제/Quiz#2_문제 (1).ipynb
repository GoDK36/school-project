삭제할 문자열 패턴은 다음과 같다.
* 플러스(+) 기호
* 슬래쉬(/) 기호로 시작하는 형태소 태그  예) /NNG, /VV

삭제하는 부분은 'rm_str'이라는 이름의 함수로 작성하여 사용할 것
______________________
결과 예시
> 너01 의 마음속 의 블루진 을 찢다 어라 솔직하다 아야 하다01 ㄴ다 .
>
> 솔직하다 아야 나03 의 영혼02 과 만나다 ㄹ 수02 있다01 다 .
>
> 그러나 솔직하다 기 가 어렵다 다 .
_________________

```python
# 함수 사용 관련 예시

# a와 b를 더한 값을 출력하는 코드
a = 1
b = 2
res = a+b
print(res)

# 더하는 부분을 'add'라는 이름의 함수로 작성하여 사용하면
def add(a, b):
    res = a+b
    return res
a = 1
b = 2
res = add(a,b)
print(res)
```
