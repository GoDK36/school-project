```python
import pandas as pd
import numpy as np
from pandas import DataFrame as df
```

```python
data = pd.read_excel(r'E:\Develop\python\school-project\언통\algorithm1.xlsx', encoding = 'utf8')
print(data)
```

```python
town_list = data['town'].tolist()

town_df = pd.DataFrame(town_list)

town_df = town_df.fillna(method='ffill')

print(town_df)

thop_list = data['thop'].tolist()
dong_list = []
for k in thop_list:
    dong = k.split(',')
    dong_list.append(dong)
print(dong_list)
# print(thop_list)
```

```python
split = data.thop.str.split(',')
split = split.apply(lambda x:pd.Series(x))
split.stack()
split.stack().reset_index(level=1,drop=True)
new = split.stack().reset_index(level=1,drop=True).to_frame('thop')
new['town'] = np.nan
new = new[['town', 'thop']]
print(new)
```

```python
merge_df = pd.merge(town_df, new, left_index=True, right_index=True)
del merge_df['town']
merge_df.columns = ['town', 'thop']
print(merge_df)
```

```python
merge_df.to_csv(r"E:\Develop\python\school-project\언통\result_김신우.csv", index=False, encoding='cp949')
```
