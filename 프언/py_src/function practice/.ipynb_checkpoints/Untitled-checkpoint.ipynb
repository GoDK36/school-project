```python
import re, os
```

```python
inDir = r"D:\GOD's folder\자료\대학교\언어학\프로그래밍 연습\교수님 자료\regex_sample"
```

```python
with open(os.path.join(inDir, 'Sejong.txt'), 'r', encoding='utf16') as inF:
    txt = inF.read()
```

```python
re.findall('<.*?>', txt)[4]
```

```python
re.findall('<title>(.*?)</title>', txt)
```

```python
s_desc = re.findall('<sourceDesc>(.*?)</sourceDesc>' ,txt, re.S)[0]
```

```python
re.findall('<title>(.*?)</title>', s_desc)[0]
```

```python
raw_ej_l = re.findall('\S*?\t(\S*?)\t.+\n', txt)
print(raw_ej_l)
```

```python
''.join(raw_ej_l)
```
